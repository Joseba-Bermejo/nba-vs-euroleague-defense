#### Question 1: What are the YoY trends in average 3PA and 3P% for both leagues independently, and compared to each other?

```python
# NBA: YoY trend in average 3PA and 3P%

ax1 = sns.barplot(data=df, x="season", y="nba_3PA", label="NBA 3PA")
sns.lineplot(data=df, x="season", y="nba_3P%", label="NBA 3P%")
ax1.bar_label(ax1.containers[0])
plt.xticks(rotation=45)
```

```python
# EuroLeague: YoY trend in average 3PA and 3P%

ax2 = sns.barplot(data=df, x="season", y="eur_3PA", label="Euro 3PA")
sns.lineplot(data=df, x="season", y="eur_3P%", label="Euro 3P%")
ax2.bar_label(ax2.containers[0])
plt.xticks(rotation=45)
```
